### [EN]

Calculator App

In this project, you will be building a basic calculator using EasyGui library. The calculator will be able to perform four basic mathematical operations: addition, subtraction, multiplication, and division.

Instructions:

1. Use the easygui module to create a GUI for the calculator.
2. Prompt the user to enter the first number in a pop-up box.
3. Provide a list of four mathematical operations to choose from in a pop-up box.
4. Prompt the user to enter the second number in a pop-up box.
5. Convert the user inputs to float
6. Perform the selected mathematical operation on the two numbers.
7. Display the result in a pop-up box.


### [AZ]

Kalkulyator proqramı

Bu proyektdə EasyGui kitabxanasını işlədərək sadə kalkulayor proqramı yazacayıq. Bu kalkulyator toplama, çıxma, vurma və bölmə əməliyyatlarını icra edə biləcək.

Təlimatlar:

1. EasyGUI kitabxanasını istifadə edərək kalkulyatorun istifadəçi interfeysini (GUI) yarat
2. Pop-up pəncərə açaraq istifadəçidən bir say alın
3. Sonra dörd riyazi əməliyyatı istifadəçiyə göstərib, birini seçməyini istəyin
4. Daha sonra bir pop-up pəncərə daha açaraq istifadəçidən ikinci bir say alın
5. Hər iki sayı "float" dəyişən tipinə çevirin
6. Bu iki ədəd arasında seçilən riyazi əməliyyatı yerinə yetirin
