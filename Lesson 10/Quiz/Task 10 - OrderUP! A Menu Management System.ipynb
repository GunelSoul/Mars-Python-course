OrderUP! A Menu Management System

You are going to create a program that will help a restaurant manage their menu items and orders. The program should use lists, sets, while loops, functions, parameters, and dictionaries.

First, create a list called "menu" that contains the following items: "hamburger", "fries", "hot dog", "pizza", "chicken sandwich", and "taco".

Next, create a set called "available_items" that contains the items "hamburger", "fries", "hot dog", and "chicken sandwich".

Now, create a function called "show_menu" that takes the "menu" list as a parameter and prints out each item in the list, one item per line.

Use a dictionary called "orders" to keep track of the orders. The keys of the dictionary should be the items ordered and the values should be the quantity of each item ordered. For example, if the user orders 2 hamburgers, the dictionary should contain the key "hamburger" with the value 2.

Create a while loop that displays a menu and prompts the user to enter their order. If the user enters an item that is not in the "available_items" set, display an error message and prompt them again. If the order is in the "available_items" set, ask the user how many they want. Then add order and its quantity to the "orders" dictionary. If the user enters "done", exit the loop.

Create a function called "show_order" that takes the order dictionary as a parameter and prints out each item ordered and the quantity of each item, one item per line.
